```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='green',linewidth=5, linestyle='dashed',marker='o', markerfacecolor='red', markersize= 12)
```

```python
days= [1,2,3,4,5,6,7]
max_t = [51,42,63,44,55,46,57]
min_t = [50,51,52,48,47,49,46]
avg= [45,45,67,23,56,34,67]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label= "max")
plt.plot(days,min_t, label= "min")
plt.plot(days,avg, label="avg")


plt.legend(loc='best', shadow=True, fontsize='small')
plt.grid()
```

```python
company=['google','amazon','fb','msft']
revenue=[90,136,89,27]
profit=[40,2,34,12]
xpos=np.arange(len(company))
xpos
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("sugar range")
plt.ylabel("Total No. Of Patient")
plt.title("Blood Sugar Level")
plt.hist([blood_sugar_men,blood_sugar_women] , bins=[80,100,125,150], rwidth=0.95, color=['y','g'],label=['men','women'])
plt.legend()
```
